```python
# One-neuron design with different M values,  # Power
import matplotlib.pyplot as plt
import numpy as np

# 数据
categories = ['N=3', 'N=8', 'N=16']  # 分类
values1 = [0.124, 0.131, 0.146]  # 第一组数据
values2 = [0.128, 0.142, 0]  # 第二组数据
values3 = [0, 0, 0.222]  # 第三组数据

# 定义每个柱状图的宽度
bar_width = 0.25

# 生成 x 轴坐标
x = np.arange(len(categories))

# 绘制柱状图
plt.bar(x - bar_width, values1, width=bar_width, label='M=3')  # 第一组柱状图
plt.bar(x, values2, width=bar_width, label='M=8')  # 第二组柱状图
plt.bar(x + bar_width, values3, width=bar_width, label='M=16')  # 第三组柱状图


# 设置标题和标签
plt.title('Comparison of Power with different M values')  # 设置标题
plt.xlabel('N')  # 设置 x 轴标签
plt.ylabel('Power/mW')  # 设置 y 轴标签
plt.xticks(x, categories)  # 设置 x 轴刻度标签为分类

# 显示图例
plt.legend()

# 显示图形
plt.tight_layout()  # 自动调整布局，避免标签重叠
plt.show()
```
