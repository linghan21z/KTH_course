```python
#3 design , N=8, ReLu,   varying precision
#Comparison of Power in different design and precision

import matplotlib.pyplot as plt
import numpy as np

# 数据
categories = ['Serial', 'Semi-parallel', 'Parallel']  # 分类
values1 = [0.32 , 1.47, 2.16]  # 第一组数据
values2 = [0.61, 2.99, 7.63]  # 第二组数据
values3 = [1.59, 6.72, 25.85]  # 第三组数据

# Design               & Control Area & Data Area & Total Area &  Power & Time Slack\\ \hline
# Serial implementation & 293.5296 & 1995.8400 &2782.0800  &0.3175mW &0.30 \\ \hline
# Semi-parallel implementation & 231 & 4319 & 8500 & 1.4708 mW & 0.00 \\\hline
# Parallel implementation & 12 & 638 & 829 & 5.2709 mW & 18.559 ns\\\hline


# 定义每个柱状图的宽度
bar_width = 0.25

# 生成 x 轴坐标
x = np.arange(len(categories))

# 绘制柱状图
plt.bar(x - bar_width, values1, width=bar_width, label='8-bit')  # 第一组柱状图
plt.bar(x, values2, width=bar_width, label='16-bit')  # 第二组柱状图
plt.bar(x + bar_width, values3, width=bar_width, label='32-bit')  # 第三组柱状图


# 设置标题和标签
plt.title('Comparison of Power in different design and precision')  # 设置标题
plt.xlabel('different design')  # 设置 x 轴标签
plt.ylabel('Power/mW')  # 设置 y 轴标签
plt.xticks(x, categories)  # 设置 x 轴刻度标签为分类

# 显示图例
plt.legend()

# 显示图形
plt.tight_layout()  # 自动调整布局，避免标签重叠
plt.show()
```
